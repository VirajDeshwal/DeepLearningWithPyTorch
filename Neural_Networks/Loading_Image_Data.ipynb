```python
import torch
from torchvision import transforms, datasets
import helper
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_transform = transforms.Compose([transforms.RandomRotation(30),
                               transforms.RandomResizedCrop(100),
                               transforms.RandomHorizontalFlip(),
                               transforms.ToTensor()])

test_transform = transforms.Compose([transforms.RandomRotation(30),
                               transforms.RandomResizedCrop(100),
                               transforms.RandomHorizontalFlip(),
                               transforms.ToTensor()])

trainset = datasets.ImageFolder(root='Cat_Dog_data/train/',transform=train_transform )
trainLoader = torch.utils.data.DataLoader(dataset=trainset, batch_size=64, shuffle=True)


testset = datasets.ImageFolder(root='Cat_Dog_data/test/', transform=test_transform)
testLoader = torch.utils.data.DataLoader(dataset=testset, batch_size=64)
```

```python
images, labels = next(iter(trainLoader))

fig, axes = plt.subplots(figsize=(10,4), ncols=4)
for i in range(4):
    ax = axes[i]
    helper.imshow(images[i], ax=ax, normalize=False)
```
